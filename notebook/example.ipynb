```python
import numpy as np
import matplotlib.pyplot as plt
import pickle
import copy
import time

from gprubin import interpolateOverDefectsGP
```

```python
pkl = 'data.pkl'
dic = pickle.load(open(pkl, 'rb'))

image = copy.deepcopy(dic['maskedImage'])

start = time.time()
interpolateOverDefectsGP(image, 15., ["SAT", "BAD"], method="jax", bin_spacing=25, 
                         threshold_dynamic_binning=2000, threshold_subdivide=20000)
end = time.time()

print(end - start)
plt.figure(figsize=(12,8))
plt.imshow(image.getImage().array, vmin=400, vmax=800, cmap=plt.cm.Greys_r)
plt.gca().invert_yaxis()
```
